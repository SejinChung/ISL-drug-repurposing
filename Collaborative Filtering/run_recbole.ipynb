```python
import torch
import pandas as pd
from recbole.config import Config
from recbole.data import create_dataset, data_preparation
from recbole.model.general_recommender import CDAE
from recbole.data.interaction import Interaction

# 1. 설정
parameter_dict = {
    'model': 'CDAE',
    'dataset': 'ls',
    'data_path': '../../data/',
    'USER_ID_FIELD': 'pubchemID',
    'ITEM_ID_FIELD': 'MeSH',
    'RATING_FIELD': 'Score',
    'load_col': {
        'inter': ['pubchemID', 'MeSH', 'Score']
    }
}

config = Config(model='CDAE', dataset='ls', config_dict=parameter_dict)

# 2. 데이터셋 로드 및 분할
dataset = create_dataset(config)
train_data, valid_data, test_data = data_preparation(config, dataset)

# 3. 모델 생성 및 불러오기
model = CDAE(config, train_data.dataset).to(config['device'])

model_path = 'saved/CDAE-Jul-11-2025_11-59-49.pth'  # 경로 조정 필요
checkpoint = torch.load(model_path, map_location=config['device'])
model.load_state_dict(checkpoint['state_dict'])
model.eval()

# 4. 토큰 -> ID 매핑
user_token2id = dataset.field2token_id[config['USER_ID_FIELD']]  # token → id
item_token2id = dataset.field2token_id[config['ITEM_ID_FIELD']]  # token → id

# 예측할 유저 및 아이템 리스트
target_pathway_merged_df = pd.read_csv('/Users/a82108/Documents/Yonsei/3-1/데마/ISL Drug Repositioning/disgenet 원본/target_pathway_merged_matrix(#2).csv', index_col=0)
target_pathway_merged_df
user_list = list(target_pathway_merged_df['pubchemID'])
item_list = list(target_pathway_merged_df.columns)[1:]

# 5. 예측 매트릭스 생성
pred_matrix = pd.DataFrame(index=user_list, columns=item_list)

# 예측 루프
for user_token in user_list:
    for item_token in item_list:
        u_id = user_token2id.get(str(user_token))
        i_id = item_token2id.get(str(item_token))

        if u_id is not None and i_id is not None:
            # RecBole 형식의 Interaction 객체 생성
            input_dict = {
                config['USER_ID_FIELD']: torch.tensor([u_id]),
                config['ITEM_ID_FIELD']: torch.tensor([i_id])
            }
            interaction = Interaction(input_dict).to(config['device'])

            with torch.no_grad():
                score = model.predict(interaction).item()

            pred_matrix.loc[user_token, item_token] = score
        else:
            pred_matrix.loc[user_token, item_token] = None
            print(f"ID not found: user={user_token}, item={item_token}")

# 6. 결과 저장
output_path = 'CF_prediction_matrix(#4).csv'
pred_matrix = pred_matrix.astype(float)
pred_matrix.to_csv(output_path)

print(f"예측 결과 저장 완료: {output_path}")
```
